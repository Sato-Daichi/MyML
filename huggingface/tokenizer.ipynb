```python
# インポート
from pprint import pprint
from transformers import AutoTokenizer
```

### __tokenizer__

```python
# トークナイザのload
# AutoTokenizerを使えば、指定したモデル名で自動的にTokenizerを読み込んでくれる

model_name = 'cl-tohoku/bert-base-japanese-whole-word-masking'
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# BertJapaneseTokenizerで読み込んでも同じ
tokenizer.__class__.__mro__
```

### __\_\_call\_\_メソッド__

```python
texts = [
    "新時代はこの未来だ",
    "世界中全部変えてしまえば変えてしまえば"
]

text = texts[0]
```

```python
# __call__メソッド
# 基本的にはこれを使用する
# 特殊トークンは自動的に追加される

tokenizer(text)
```

```python
# 複数の文章を入れても良い

tokenizer(texts)
```

```python
# オプション引数をつける

tokenizer(
    text,
    padding="max_length", # "max_length" or "longest"
    truncation=True, # 切り捨て
    max_length=10, # padding="max_length" と truncation=Trueに影響
    return_tensors="pt" # 返り値のデータ型
)
```

### __その他のメソッド__

```python
# テキスト→トークン

tokens = tokenizer.tokenize(text)
tokens
```

```python
# トークン→テキスト

tokenizer.convert_tokens_to_string(tokens)
```

```python
# トークン→トークンID

ids = tokenizer.convert_tokens_to_ids(tokens)
ids
```

```python
# トークンID→トークン

tokenizer.convert_ids_to_tokens(ids)
```

```python
# テキスト→トークン→トークンID
# 自動的に特殊トークンが入る

ids = tokenizer.encode(text)
ids
```

```python
# トークンID→トークン→テキスト

tokenizer.decode(ids)
```
