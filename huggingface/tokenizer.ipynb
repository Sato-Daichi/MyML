```python
# インポート
from pprint import pprint
from transformers import AutoTokenizer
```

```python
# トークナイザを読み込み
# AutoTokenizerを使えば、指定したモデル名で自動的にTokenizerを読み込んでくれる

model_name = 'cl-tohoku/bert-base-japanese-whole-word-masking'
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
texts = [
    "新時代はこの未来だ",
    "世界中全部変えてしまえば変えてしまえば"
]
```

```python
# __call__メソッド
# 基本的にはこれを使用する

tokenizer(texts)
```

```python
# オプション引数をつけて__call__メソッド

tokenizer(
    texts,
    padding="max_length", # "max_length" or "longest"
    truncation=True, # 切り捨て
    max_length=10, # padding="maxx_length" と truncation=Trueに影響
    return_tensors="pt" # 返り値のデータ型
)
```

```python
# テキスト→トークン
text = "新時代はこの未来だ"

tokens = tokenizer.tokenize(text)
tokens
```

```python
# トークン→ID

tokenizer.convert_tokens_to_ids(tokens)
```

```python
# テキスト→トークン→ID

ids = tokenizer.encode(text)
ids
```

```python
# ID→トークン→テキスト

tokenizer.decode(ids)
```

```python
# トークンとIDの対応関係

tokenizer.get_vocab()
```

```python
# 特殊トークンのID

tokenizer.pad_token_id, \
tokenizer.unk_token_id, \
tokenizer.cls_token_id, \
tokenizer.sep_token_id, \
tokenizer.mask_token_id
```

```python
# 全ての特殊トークンとID
tokenizer.all_special_tokens, \
tokenizer.all_special_ids
```
