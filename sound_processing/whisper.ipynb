#### __whisperコマンドで実行__

__基本的な使い方__

```
whisper ファイル名
```

__オプション引数__

- `--model モデル名`

    - モデルを指定

    - デフォルトは`small`

    - `tiny`, `base`, `small`, `medium`, `large`がある

- `--language 言語名`

    - 言語を指定

    - 言語を指定しなくても自動で言語を推測してくれる

    - 日本語は`ja`または`Japanese`

- `--task translate/transcribe`

    - タスクを指定

    - `transcribe`で音声書き起こし

    - `translate`で英語に翻訳

- `--help`

    - オプション一覧を表示



````python
# 言語を指定しない場合

!whisper $file_name
````

````python
# 言語を指定した場合

!whisper $file_name --language ja
````

````python
# 日本語から英語に翻訳

!whisper $file_name --language Japanese --task translate
````

````python
# オプション一覧

!whisper --help
````

#### __Pythonで実行__

````python
import whisper
````

````python
# モデルのロード

model = whisper.load_model("medium")
model.device
````

````python
# 音声書き起こし

result = model.transcribe(
    file_name,
    verbose=True,
    language="ja"
)
````

````python
# 辞書で返ってくる
# text: 書き起こしたテキスト
# segments: 何秒から何秒まで何を話したか
# language: 言語

result.keys()
````

````python
result["text"]
````

````python
# segmentを整形

import pandas as pd
pd.DataFrame(result["segments"])[["id", "start", "end", "text"]]
````

````python
# 英語に翻訳

result = model.transcribe(
    file_name,
    verbose=True,
    language="ja",
    task="translate",
)

result["text"]
````
