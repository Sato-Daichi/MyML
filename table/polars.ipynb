```python
# DataFrameの作成①
pl.DataFrame([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])
```

```python
# DataFrameの作成②
pl.DataFrame({
    "整数": [1, 2, 3, 4],
    "ひらがな": ["あ", "い", "う", "え"]
})
```

```python
# 条件式
df.filter((pl.col("longitude") > -120) & (pl.col("latitude") < 35))
```

```python
# 列を選択
# 1列のみ選択してもDataFrameとなる
df.select(["longitude", "latitude"])
```

```python
# エクスプレッションのリストを指定することも可能
# スカラー値はブロードキャストされる
# 列名は自動的に付与される
df.select([
    pl.col("longitude") * 2,
    pl.col("latitude").mean(), # スカラー値
    pl.col("housing_median_age") + pl.col("total_rooms"), # 列名に注意
])
```

```python
# 列を選択
# 1列のみ選択するとSeriesとなる
df.get_column("longitude")
```

```python
# 重複行の削除
# pandasのdrop_duplicates()と同じ
df.unique()
```

```python
# データ型で検索
df.select(pl.col(pl.Float64))
```

```python
# 全ての列を選択
# pl.col("*")も可
df.select(pl.all())
```

```python
# 列を除外して取得
df.select(pl.exclude("longitude"))
```

```python
# Seriesのリストに変換
df.get_columns()
```

```python
# その列が何番目の列か
df.find_idx_by_name("latitude")
```

```python
# 列の追加(更新)①
# 列名を変更しない場合は列が上書きされる
# aliasで列名を変更
df.with_columns(
    (pl.col("longitude") * 2).alias("longitude2")
)
```

```python
# 列の追加(更新)②
# キーワード引数でも指定できる
df.with_columns(
    longitude_mean=pl.mean("longitude"),
    longitude_max=pl.max("longitude"),
)
```

```python
# 列の追加③
# selectを下のように使えば列を追加できる
df.select([
    pl.col("*"), # 全ての列
    pl.lit(0).alias("zero"), # 追加する列
])
```

```python
# 左端に0始まりの整数列「row_nr」を作成
# indexのような役割を果たせる
df.with_row_count()
```

```python
# ソート①
# エクスプレッション
df.select(pl.col("longitude").sort())
```

```python
# sum(行方向)
df.select(pl.sum(["longitude", "latitude"]))
```

```python
# sum(列方向)
df.select(pl.col(["longitude", "latitude"]).sum())
```

```python
df.select([
    pl.count("longitude").alias("count"),
    pl.n_unique("longitude").alias("n_unique"),
    pl.max("longitude").alias("max"),
    pl.min("longitude").alias("min"),
    pl.mean("longitude").alias("mean"),
    pl.median("longitude").alias("median"),
    pl.std("longitude").alias("std"),
    pl.var("longitude").alias("var"),
])
```

```python
# groupbyはoverで書ける
df.select(pl.col("latitude").sum().over("longitude"))
```

```python
# 条件分岐
# when→then→otherwise
df.select(
    pl.when(pl.col("longitude") > -120)
    .then("a")
    .when(pl.col("longitude") > -115)
    .then("b")
    .otherwise("c")
)
```

```python
# DataFrameの結合
a, b = df[:10], df[10:20]
pl.concat([a, b])
```

```python
# groupby
# 行はランダムにソートされる
df.groupby("longitude").mean()
```

```python
# GroupByオブジェクトをイテレータとして使う
# group_name, group_dfが返る
grouped = df.groupby("longitude")

for group_name, group_df in grouped:
    print(group_name)
    print(group_df)
    break
```

```python
# 欠損値補完
# nullとNanでそれぞれメソッドがある
# 平均値補完は"mean"
# 他にも"backward", "max", "zero"などがある

# df.fill_nan(0)
df.fill_null(0)
```

```python
# 文字列処理
# strアクセサ
text_df = pl.DataFrame({
    "number": ["one", "two", "three", "four"]
})

text_df
```

```python
text_df.with_columns([
    pl.col("number").str.lengths().alias("length"), # 長さ(バイト数)
    pl.col("number").str.to_uppercase().alias("upper"), # 大文字
    pl.col("number").str.slice(1, 3).alias("slice"), # スライス
    pl.col("number").str.n_chars().alias("n_chars"), # 長さ(全角文字等を使う場合)
    pl.col("number").str.contains("o").alias("contain_o"), # 文字列を含むか
    pl.col("number").str.replace("o", "O").alias("replace"), # 置換
])
```
