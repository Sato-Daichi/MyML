### __確率分布のグラフを描画する__

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
````

### __scipy.statsについて__

各確率分布で共通のメソッドを持っているので、

```py
scipy.stats.確率分布名.メソッド
```

の形で使える。

|確率分布名|モジュール名|
|-------------|---------------|
|正規分布|norm|
|カイ2乗分布|chi2|
|F分布|f|

|意味|英語|メソッド名|
|------|----|-----------|
|確率密度関数|Probability Density Function|pdf()|
|累積分布関数|Cumulative Distribution Function|cdf()|

### __正規分布__

````python
r = 40
x = np.arange(-r, r, 0.1)

mean = 0
std = 10
y = stats.norm.pdf(x, mean, std) # 平均、標準偏差

plt.plot(x, y)
plt.show()
````

````python
# 積分区間を表示

r = 40
x = np.arange(-r, r, 0.1)
y1 = np.zeros(len(x)) # x軸

mean = 0
std = 10
y2 = stats.norm.pdf(x, mean, std) # 平均、標準偏差

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y1, c="w") # x軸
ax.plot(x, y2)

# 2つの関数で囲まれた領域を塗りつぶす
a, b = 0, 10
ax.fill_between(x, y2, y1, facecolor='blue', alpha=0.2, where=(a<x) & (x<b))
plt.show()
````

### __カイ2乗分布__

````python
r = 20
x = np.arange(0, r, 0.1)

df = 3 # 自由度 (degree of freedom)
y = stats.chi2.pdf(x, df) # 平均、標準偏差

plt.plot(x, y)
plt.show()
````

````python
# 自由度ごとに表示
r = 20
x = np.arange(0, r, 0.1)

dfs = [1, 2, 3, 5, 10]
for df in dfs:
    y = stats.chi2.pdf(x, df) # 平均、標準偏差
    plt.plot(x, y, label=f"df {df}")

plt.legend()
plt.show()
````

````python
# 積分区間を表示

r = 20
x = np.arange(0, r, 0.1)
y1 = np.zeros(len(x)) # x軸

df = 3
y2 = stats.chi2.pdf(x, df) # 平均、標準偏差

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y1, c="w") # x軸
ax.plot(x, y2)

# 2つの関数で囲まれた領域を塗りつぶす
a, b = 5, 10
ax.fill_between(x, y2, y1, facecolor='blue', alpha=0.2, where=(a<x) & (x<b))
plt.show()
````
